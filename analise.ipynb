```python
#Faz imports necessários para análise
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Abre arquivo csv em Dataframe pandas
df = pd.read_csv("dados_alfabetizacao.csv")
```

### Quais perguntas poderamos fazer aos nossos dados? 

Tendo em vista que temos os números de alfabetizados e não alfabetizados por município e por UF, algumas perguntas podem ser respondidas, como: 

1 - Qual o percentual de Jovens alfabetizados X não alfabetizados no brasil? 

2 - Qual o percentual de Jovens alfabetizados X não alfabetizados no município de Joanópolis SP?

```python
# Pergunta 1 - Qual o total de alfabetizados e não alfabetizados no brasil?
# Sumarizamos o total de alfabetizados e não alfabetizados por município e plotando

#Cria itens para o plot
items = ['Jovens Alfabetizados', 'Jovens não Alfabetizados']

#Faz sumarização no dataframe definindo proporções
proportions = [int(df['sim'].sum()), int(df['nao'].sum())]

colors = ['g', 'r']

plt.pie(proportions, labels=items, colors=colors,
        startangle=20, shadow=True, explode=(0.1, 0),
        radius=1.2, autopct='%1.1f%%', )


plt.title('Jovens alfabetizados X não alfabetizados no brasil')
plt.legend()
plt.show()
```

```python
# Pergunta 2 - Qual o percentual de Jovens alfabetizados X não alfabetizados no município de Joanópolis SP?

#Seta massa para o município específico
df_joa = df[df['nome'] == 'Joanópolis (SP)']

# Sumarizamos o total de alfabetizados e não alfabetizados por município e plotando

#Cria itens para o plot
items = ['Jovens Alfabetizados', 'Jovens não Alfabetizados']

#Faz sumarização no dataframe definindo proporções
proportions = [int(df_joa['sim'].sum()), int(df_joa['nao'].sum())]

colors = ['g', 'r']

plt.pie(proportions, labels=items, colors=colors,
        startangle=20, shadow=True, explode=(0.1, 0),
        radius=1.2, autopct='%1.1f%%', )


plt.title('Jovens alfabetizados X não alfabetizados Joanópolis SP')
plt.legend()
plt.show()
```
